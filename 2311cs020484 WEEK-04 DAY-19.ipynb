```python
data=sns.load_dataset('iris')
data
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data['species']=lb.fit_transform(data['species'])
```

```python
data.head(3)
```

```python
clf=IsolationForest(random_state=500)
clf.fit(data)
```

```python
clf.predict(data) #It will give some values 
#1 -data
#-1 - outlier data
```

```python
clf=IsolationForest(random_state=500,contamination=0.01)
clf.fit(data)   #we have the hyper - parameter for smart machine elarning alogo it is contaminator
#This hyper-parameter is a trial and error method
```

```python
clf.predict(data) #now after usin the contaminator parameter the outliers are reduced
```

```python
y_pred_outliers=clf.predict(data)
y_pred_outliers
```
